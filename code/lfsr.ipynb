```python
import numpy as np
```

```python
lfsr = {
    "degree": 6,
    "taps": [0, 5],
    "initial_state": np.array([1, 1, 1, 1, 1, 1]),
    "exp_period": 2**6 - 1,
    "states": set(),
    "output_bits": []
}

state = lfsr["initial_state"]

def new_bit(state, lfsr):
    # XOR all values in the tap positions that are 1
    tapped_values = state[lfsr["taps"]]
    print("Tapped values:", tapped_values)
    
    xor_result = np.bitwise_xor.reduce(tapped_values)
    
    print("XOR result:", xor_result)
    
    return xor_result

def shift(state, lfsr):
    state_str = np.array2string(state)
    lfsr["states"].add(state_str)
    bit = new_bit(state, lfsr)
    state[1:] = state[:-1]  # Shift all bits to the right
    state[0] = bit
    return state


def output_bits(state,lfsr):
    exp_period = lfsr["exp_period"]
    print("exp_period", exp_period)
    for i in range(exp_period):
        print("-------")
        output_bit = state[-1]
        
        lfsr["output_bits"].append(output_bit)

        print(state, "--->", output_bit)

        state = shift(state, lfsr)
        if np.array2string(state) in lfsr["states"]:
            print(f"Initial state repeated after {i} iterations")
            break

output_bits(state,lfsr) 
print(lfsr["output_bits"][:10])

```

```python
lfsr = {
    "degree": 8,
    "taps": [0, 4, 5, 6],
    "exp_period": 2**8 - 1,
    "initial_state": np.array([0, 1, 0, 1, 0, 1, 0, 1]),
    "states": set(),
    "output_bits": []
    }

state = lfsr["initial_state"]

output_bits(state,lfsr)
print(lfsr["output_bits"][:20])
```
