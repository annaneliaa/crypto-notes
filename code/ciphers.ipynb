```python
# works only on letters 'a' to 'z'
# shift using unicode values
def cshift(text, shift):
    result = []
    for char in text:
        if char.isalpha():
            if char.islower():
                result.append(chr((ord(char) - 97 + shift) % 26 + 97))
            else:
                result.append(chr((ord(char) - 65 + shift) % 26 + 65))
        else:
            result.append(char)
    return ''.join(result)

def shiftback(text, shift):
    return cshift(text, -shift)
```

```python
with open('cipher.txt') as f:
    cipher = f.read().strip()

for i in range(26):
    print(i, shiftback(cipher, i))
```

```python
def alphabet_loc(char):
    return ord(char) - 97

# has bugs?
# vary the length of repeating sequence in the text
def find_repeat(text, seq_len):
    for i in range(len(text) - seq_len):
        seq = text[i:i+seq_len]
        for j in range(i+seq_len, len(text) - seq_len):
            if text[j:j+seq_len] == seq:
                return seq, i, j
    return None

# find all indexes of substring in text
def find_index(text, sub):
    indexes = []
    for i in range(len(text) - len(sub)):
        if text[i:i+len(sub)] == sub:
            indexes.append(i)

    return indexes
```
